MONTE CARLO SIMULATION TASK

QUESTION:

Show the trajectory of an electron motion in 2D using Monte Carlo Simulation, and hence determine the angle, phi (φ) and the mean free path, lambda (Λ) for: ten (10), hundred (100), a thousand (1000) and ten thousand (10000) collisions.

Given that:

Phi (φ) = 2πa, a --- [0,1] "random valuable"

Lambda (Λ) = -λ.ln(a)

NB: The value of "a" must be selected differently for each collision(s).

where: Phi, φ - is the angle between the electron collisions with other particles, a - is Bohr radius, lambda, Λ - is the mean free path, λ - is lambda null which is assumed to be 1, ln(a) - is the natural logarithsm of a in base e.

Simulation for ten (10) Collisions

```python
#@title
import numpy as np
import matplotlib.pyplot as plt
import random

numbere = 1
num_a = 10
e = []
for j in range(numbere):

  a = np.random.uniform(size=num_a)
  b = np.random.uniform(size=num_a)
  ex = [0]
  ey = [0]
  for i in range(len(a)):
    Lambda = -np.log(b[i])
    phi = 2*np.pi*a[i]

    ex.append(Lambda*np.cos(phi))
    ey.append(Lambda*np.sin(phi))
  plt.plot(ex, ey, linestyle='-',color ="navy", linewidth = 3, marker='o', markerfacecolor='green', markersize=10)
plt.show()

# Graph of the simulation
L = []
a = np.random.uniform(size=num_a)
b = np.random.uniform(size=num_a)
ex = [0]
ey = [0]

for i in range(len(a)):
  Lambda = -np.log(b[i])
  phi = 2*np.pi*a[i]

  ex.append(Lambda*np.cos(phi))
  ey.append(Lambda*np.sin(phi))
lx = 0
ly = 0
N = []
n = 1
for i in range(len(a)):
  N.append(n)
  n = n + 1
  lx = lx + ex[i]
  ly = ly + ey[i]
  L.append(lx+ly)

plt.plot(N, L, color ="darkblue")
plt.title("The Plot of Mean Free Path (Length) against Number of Collisions")
plt.xlabel("Number of Collisions")
plt.ylabel("Mean Free Path (Length)")
plt.show()

# Absolute value of the graph
L = []
a = np.random.uniform(size=num_a)
b = np.random.uniform(size=num_a)
ex = [0]
ey = [0]

for i in range(len(a)):
  Lambda = -np.log(b[i])
  phi = 2*np.pi*a[i]

  ex.append(Lambda*np.cos(phi))
  ey.append(Lambda*np.sin(phi))
lx = 0
ly = 0
N = []
n = 1
for i in range(len(a)):
  N.append(n)
  n = n + 1
  lx = lx + abs(ex[i])
  ly = ly + abs(ey[i])
  L.append((lx+ly)**(1/2))

plt.plot(N, L, color ="darkblue")
plt.title("The Plot of Mean Free Path (Length) against Number of Collisions: Absolute Values")
plt.xlabel("Number of Collisions")
plt.ylabel("Mean Free Path (Length)")
plt.show()
```
